```python
#SPecify the datafolder and the path containing your files
DATADIR= r"C:\Christine\Hydrology\Gridded data Observed Finland\Daily H/*nc"
path= glob(f'{DATADIR}')
```

```python
# Create an empty dataframe to hold all the results

df= pd.DataFrame()

for file in path:
    try:

        ds= xr.open_dataset(file) # open the .nc files
        #rename the vairables to shorter names
        #ds= ds.rename({})

        #Use the line code for creating a dataframe
        df1= ds.Rh.sel(Lon=400000, Lat= 7200000, method='nearest').to_dataframe()

        #Dropping extra columns
        #df1= df1.drop(['Lon','Lat'], axis=1)

        # Append the DataFrame to the existing DataFrame
        df_appended = pd.concat([df1, df])
        df= df_appended
        
    except Exception as err:
        print(err)
        
    
    
```
